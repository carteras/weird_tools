```python
students = pd.read_csv(data / "acs-Student Reports-Academic Record.2024.S2.AI1.csv", converters={'Student ID': '{:0>7}'.format})
```

```python
students.CoursewithOtherSchool1.unique()
```

```python
networking_students = students[students['CoursewithOtherSchool1'] == 'NETWORKING AND SECURITY ']
```

```python
networking_students.columns
```

```python
# Group by 'Fullname1' and 'YearLevel1', and apply the list operation on 'UnitwithOtherCheck1'
result = networking_students.groupby(['Fullname1', 'YearLevel1'])['UnitwithOtherCheck1'].apply(list).reset_index()

# Sort the result by 'YearLevel1'
result_sorted = result.sort_values(by='YearLevel1')

all_units = []

# Display the sorted result
for idx, row in result_sorted.iterrows():
    # print(row['YearLevel1'], row['Fullname1'])
    for unit in row['UnitwithOtherCheck1']:
        unit = unit.strip()
        # print(f"\t{unit}")
        if unit[-1] in 'ab': continue 
        if unit == "Information Tech - Non-ACT": continue

        if unit not in all_units: all_units.append(unit)
    # print("\t",row['UnitwithOtherCheck1'])  # Corrected column name
```

```python


for unit in all_units:
    print(unit)
```

```python
for idx, row in result_sorted.iterrows():
    print(row['YearLevel1'], row['Fullname1'])
    units_completed = []
    for unit in row['UnitwithOtherCheck1']:
        unit = unit.strip()
        units_completed.append(unit)
        print(f"\tCompleted  - {unit}")

    for unit in all_units:
        if unit in units_completed: continue
        print(f"\tIncomplete - {unit}")
```
