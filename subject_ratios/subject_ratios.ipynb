```python
from inspect import currentframe, getframeinfo
from dataclasses import dataclass
from pathlib import Path
import pandas as pd
filename = getframeinfo(currentframe()).filename
here = Path.cwd()
data = here.parent / ".data"
absences = data / "absence_details_of_students.csv"
df = pd.read_csv(absences)
```

```python
df.columns
```

```python
all_units = df['Unit Name'].unique()
```

```python
ignored_units = [
    'HeLP',
    'Futures',
    'Contemporary'
]

for unit in all_units:
    if any(forbidden_word in unit for forbidden_word in ignored_units): continue
    students_in_unit = df[df['Unit Name'] == unit]
    number_students = len(students_in_unit['Student Key'].unique())
    number_of_lines = len(students_in_unit['Unit Code'].unique())
    print(f"{unit}: students/lines {number_students}/{number_of_lines} == {number_students/number_of_lines:,.2f}")
  
```
