```python
from inspect import currentframe, getframeinfo
from dataclasses import dataclass
from pathlib import Path
import pandas as pd
filename = getframeinfo(currentframe()).filename
here = Path.cwd()
data = here.parent / ".data"
lines = data / "lines"
```

```python
all_files = [file for file in lines.glob("*") if file.is_file()]
all_files
```

```python
# Open the first file in the list
col = {
    'room_info': [],
    'course_code': [],
    'room_code': [],
    'teacher_name': [],
    'line_number': []
}
if all_files:
    file_path = all_files[7]
    for file_path in all_files:

        with open(file_path, 'r', encoding='utf-8') as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]  # Remove empty lines

        data = []
        
        room = "Room:"
        salutations = ["Mr", "Ms", "Mrs", "Mx", "Miss", "Sonali", "Flexible", 'Marina']
        pattern = r'\(([^()]+)$'
        ignore = ['Composite', '800CGS1']


        for line in lines:
            # print(line)
            if (line.startswith(tuple(ignore))): continue
            if line.startswith('Room:'): 
                col["room_code"].append(line.replace("Room: ", ""))
            elif line.startswith(tuple(salutations)):
                col['teacher_name'].append(line.strip())
            else: 
                
                line = line.replace(")", "")  # Remove the closing parenthesis

                try:
                    # rsplit with an integer 1 to split from the right only once
                    room_info, course_code = line.rsplit("(", 1)
                    
                except ValueError:
                    print("-=-=-=-=-=-", line)  # Handles cases without '('
                    room_info = line
                    course_code = ""
                # Clean up whitespace
                room_info = room_info.strip()
                course_code = course_code.strip()
                col['room_info'].append(room_info)
                col['course_code'].append(course_code)
                col['line_number'].append(course_code[0])

        


        
df = pd.DataFrame(col)
print(len(col["course_code"]), len(col["room_code"]), len(col["room_info"]), len(col["teacher_name"]), len(col['line_number']))
    
```
